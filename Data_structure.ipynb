```python
# dictionary
## generating/creating a dict (,ultiple ways to create a dictonary in python)

# Empty dict
d = {}
d = dict()

# With key-value pairs
student = {"name": "Ravi", "age": 20, "city": "Bengaluru"}

# Using dict() constructor
student = dict(name="Ravi", age=20)

# From list of tuples
pairs = [("a", 1), ("b", 2)]
d = dict(pairs)
```

```python
## Accessing dict element (access values using keys-two ways)

student = {"name": "Ravi", "age": 20}

# Using square bracket
print(student["name"])     # Ravi

# Using get() — safe, no KeyError
print(student.get("age"))   # 20
print(student.get("marks", "N/A"))  # N/A (default)
```

```python
# modifying dict elements (change a value by assinging to an existing key)

student = {"name": "Ravi", "age": 20}

student["age"] = 21         # Modify existing key
print(student)
```

```python
## Accessing all keys and values(use built in methods to view keys , values or both)

student = {"name": "Ravi", "age": 20}

print(student.keys())    
print(student.values()) 
print(student.items())   
```

```python
## update()-change value by key(changes existing values or adds new keys)

student = {"name": "Ravi", "age": 20}

student.update({"age": 22})        # Change value
student.update({"city": "Mysuru"})  # Add new key
print(student)
```

```python
## adding new key-value pairs(assign a new key directly to add it to the dict)

student = {"name": "Ravi"}

student["marks"] = 95       # Add new key
student["grade"] = "A"      # Add another key
print(student)
```

```python
## del=delete a key(del removes a specific key-value pair permanently)

student = {"name": "Ravi", "age": 20, "city": "Mysuru"}

del student["city"]
print(student)
```

```python
## pop()-remove and return (pop() removes a key and return its value)

student = {"name": "Ravi", "age": 20}

removed = student.pop("age")
print(removed)   # 20
print(student)   # {'name': 'Ravi'}

# With default (no KeyError if missing)
val = student.pop("marks", 0)
print(val)       # 0
```

```python
## Clear()-empty the dict(claer() removes all-value pairs, dict remains)

student = {"name": "Ravi", "age": 20}

student.clear()
print(student)   # {}
```

```python
## shallow copy vs deep copy (shallow copy shares nested objects;deep copy is fully independent)

import copy

original = {"name": "Ravi", "scores": [90, 85]}

# Shallow copy — nested list is shared!
shallow = original.copy()
shallow["scores"].append(95)
print(original["scores"])  

# Deep copy — fully independent
original2 = {"name": "Ravi", "scores": [90, 85]}
deep = copy.deepcopy(original2)
deep["scores"].append(95)
print(original2["scores"])  
```

```python
## Iterate over keys (loop through all keys in a dictionary)

student = {"name": "Ravi", "age": 20, "city": "Mysuru"}

for key in student:           # or student.keys()
    print(key)
```

```python
## iterate over values (loop through all values in a dictionary)

student = {"name": "Ravi", "age": 20}

for val in student.values():
    print(val)

# Iterate over key-value pairs
for key, val in student.items():
    print(key, "→", val)
```

```python
## dict comprehension (build a dictionary in one line using comprehension syntax)

# Squares of numbers 1-5
squares = {x: x**2 for x in range(1, 6)}
print(squares)


# From two lists
keys   = ["a", "b", "c"]
values = [1, 2, 3]
d = {k: v for k, v in zip(keys, values)}
print(d)
```

```python
## conditional dict comprehension  (Add ab if condition to filter item inside the comprehension)

# Only even numbers
evens = {x: x**2 for x in range(1,11) if x % 2 == 0}
print(evens)


# Students who passed (marks >= 50)
marks  = {"Ravi":80, "Anu":45, "Priya":72}
passed = {k: v for k, v in marks.items() if v >= 50}
print(passed)
```

```python
## merge 2 dict (three way to merge dictionaries)

d1 = {"a": 1, "b": 2}
d2 = {"c": 3, "d": 4}

# Method 1 — update() (modifies d1)
d1.update(d2)

# Method 2 — unpacking (Python 3.5+)
merged = {**d1, **d2}

# Method 3 — | operator (Python 3.9+)
merged = d1 | d2

print(merged)
```

```python
## extra dict methods (other useful dictionary methods you should know)

d = {"a":1, "b":2}

# setdefault() — add key only if not present
d.setdefault("c", 3)
print(d)   

# popitem() — remove and return last inserted pair
item = d.popitem()
print(item)  

# fromkeys() — create dict from list of keys
keys = ["x", "y", "z"]
new_d = dict.fromkeys(keys, 0)
print(new_d)

# len() — count of key-value pairs
print(len(d))   

# in operator — check if key exists
print("a" in d)   

# Nested dict
nested = {"student": {"name": "Ravi", "age": 20}}
print(nested["student"]["name"])  
```
