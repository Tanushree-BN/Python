```python
## Data Aggregation and Grouping
df.head()
```

```python
group_mean=df.groupby('Name')['Age'].mean()
print(group_mean)
```

```python
grouped_sum=df.groupby(['Name','Gender'])['Age'].sum()
print(grouped_sum)
```

```python
## aggregate multiple function

grouped_agg=df.groupby('Name')['value'].agg(['mean','sum','max'])
grouped_agg
```

```python
## Merging and joining dataframes
## create sample data frames
import pandas as pd
df1 = pd.DataFrame({'key':['A','B','C'],'value1':[1,2,3]})
df2 = pd.DataFrame({'key':['A','B','D'],'value2':[4,5,6]})
```

```python
df2
```

```python
df1
```

```python
## Merge Dat frames on the 'key columns'8.1-numpy.ipynb
pd.merge(df1,df2,on="key",how="inner")
```

```python
pd.merge(df1,df2,on="key",how="outer")
```

```python
pd.merge(df1,df2,on="key",how="left")
```

```python
pd.merge(df1,df2,on="key",how="right")
```
