
Standard Library overview

python's standard library is a vast collection of modules and pckages that come bundled with python, providing a wide range of functionalities out of box. here's an overview of some of the most commonly used modules and packages in the python standart library

```python
import array 
arr=array.array('i',[1,2,3,4])
print(arr)
```

```python
import math
print(math.sqrt(16))
print(math.pi)
```

```python
## random 

import random
print(random.randint(1,10))
print(random.choice(['apple','banana','cherry']))
```

```python
## file and directory access

import os
print(os.getcwd())
```

```python
## high level operations on files and collection of files

import shutil
shutil.copyfile('source.txt','destination.txt')
```

```python
## data serialization

import json
data = {'name':'Alice','age':30,'city':'New York'}

json_str = json.dumps(data)
print(json_str)
print(type(json_str))

parsed_data=json.loads(json_str)
print(parsed_data)
print(type(parsed_data))
```

```python
## CSV

import csv
with open('example.csv',mode='w',newline='') as file:
    writer=csv.writer(file)
    writer.writerow(['Name','age'])
    writer.writerow(['Alice',30])

with open('example.csv',mode='r') as file:
    reader=csv.reader(file)
    for row in reader:
        print(row)
```

```python
## date time

from datetime import datetime, timedelta
now= datetime.now()
print(now)

yesterday= now-timedelta(days=1)

print(yesterday)
```

```python
## time

import time
print(time.time())
time.sleep(2)
print(time.time())
```

```python
## regular expresition

import re
pattern=r'\d+'
text='There are 123 apples'
match=re.search(pattern,text)
print(match.group())
```
