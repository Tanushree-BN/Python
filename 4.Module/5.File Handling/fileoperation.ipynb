```python
## File operation read and write files

## File handling is a crucial part of any programming language.python provides built in function and methods read from and write to file, both text and binary.
 ##this lesson will cover the basics of filenhandling , including reading and writing text files and binary files.
```

```python
## read a whole file

with open('example.txt', 'r') as file:
    content = file.read()
    print(content)
```

```python
## Read a file line by line

with open('example.txt','r') as file:
    for line in file:
        print(line.strip())  ## strip() remove the newline character
```

```python
## writing a file

with open('example.txt','w') as file:
    file.write('Hello world!\n')
    file.write('this is a new line.')
```

```python
## write a file(with out iverwriting)

with open ('example.txt','a') as file:
    file.write("append operation taking place!\n")
```

```python
## writing a list of lines to a file

lines=['First line \n','second line \n','Third line \n ']
with open('example.txt','a') as file:
    file.writelines(lines)
```

```python
## Binary files

# writing to a binary file
data=b'\x00\x01\x02\x03\x04'
with open('example.bin','wb') as file:
    file.write(data)
```

```python
## reading a binary file

with open('example.bin','rb') as file:
    content = file.read()
    print(content)
```

```python
## read the content from text file and write to a destinatiom text file
# copying a text file

with open ('example.txt','r') as Source_file:
    content = Source_file.read()

with open('destination.txt','w') as destination_file:
    destination_file.write(content)
```
